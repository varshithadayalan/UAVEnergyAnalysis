```python
import os
import numpy as np
import pandas as pd

# Data Preprocessing
from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler

# ML Model
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

# Performance Metrics
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

#Libraries for plots
import matplotlib.pyplot as plt
import seaborn as sns
```

### Path Directory

```python
train_path = "/content/drive/MyDrive/Data Analytics - Intern/Intern - Jayasri/Dataset/Energy Consumption in drone dataset/dron_lider_train.csv"
test_path = "/content/drive/MyDrive/Data Analytics - Intern/Intern - Jayasri/Dataset/Energy Consumption in drone dataset/dron_lider_test.csv"
```

```python
X = pd.read_csv(train_path)
X = X.iloc[:,:]
X.head(), X.columns
X.shape, X.columns
```

##### since the number of test data greater than train data, reducing the test data to check the model performance

```python
Num_test_data = 3000
Y = pd.read_csv(test_path)
Y = Y.iloc[:Num_test_data,:]
Y.shape, Y.columns
```

### Train and Test Features Split

```python
input_features = ['Orientation X', 'Orientation Y', 'Orientation Z', 'Orientation W', 
                  'vx', 'vy', 'vz', 'wx', 'wy', 'wz', 'mass']
output_feature = ["Power"]
X_train = X.loc[:, input_features[:]].values
Y_train = X.loc[:,output_feature].values
X_test = Y.loc[:, input_features[:]].values
Y_test = Y.loc[:,output_feature].values
print("X_train:", X_train.shape, "Y_train:", Y_train.shape)
print("X_test:", X_test.shape, "Y_test:", Y_test.shape)
# X_train, Y_train, X_test, Y_test
```

```python
Y_train.ravel(), Y_test.ravel()
```

### ML Model Fit and predict

#### Model with original raw data

```python
model_LR = LinearRegression()
model_LR_fit = model_LR.fit(X_train, Y_train)
model_LR_pred = model_LR_fit.predict(X_test)

r2_LR = r2_score(Y_test, model_LR_pred).round(4)
print("r2 score of Linear Regression:", r2_LR)
mse_LR = mean_squared_error(Y_test, model_LR_pred).round(4)
print("mse error of Linear Regression:", mse_LR)
mae_LR = mean_absolute_error(Y_test, model_LR_pred).round(4)
print("mae error of Linear Regression data:", mae_LR)
```

```python
model_RF = RandomForestRegressor(n_estimators=200)
model_RF_fit = model_RF.fit(X_train, Y_train.ravel())
model_RF_pred = model_RF_fit.predict(X_test)

r2_RF = r2_score(Y_test.ravel(), model_RF_pred).round(4)
print("r2 score of Random Forest Regression:", r2_RF)
mse_RF = mean_squared_error(Y_test.ravel(), model_RF_pred).round(4)
print("mse error of Random Forest Regression:", mse_RF)
mae_RF = mean_absolute_error(Y_test, model_RF_pred).round(4)
print("mae error of Random Forest Regression data:", mae_RF)
```

```python
model_KNN = KNeighborsRegressor(n_neighbors=4)
model_KNN_fit = model_KNN.fit(X_train, Y_train)
model_KNN_pred = model_KNN_fit.predict(X_test)

r2_KNN = r2_score(Y_test, model_KNN_pred).round(4)
print("r2 score of KNeighborsRegressor:", r2_KNN)
mse_KNN = mean_squared_error(Y_test, model_KNN_pred).round(4)
print("mse error of KNeighborsRegressor:", mse_KNN)
mae_KNN = mean_absolute_error(Y_test, model_KNN_pred).round(4)
print("mae error of KNeighborsRegressor data:", mae_KNN)
```

#### Model with standardscalar data

```python
# Standard Scalar
scalar = StandardScaler()
X_train_scalar = scalar.fit_transform(X_train)
X_test_scalar  = scalar.fit_transform(X_test)
```

```python
model_scalar_LR = LinearRegression()
model_scalar_LR_fit = model_scalar_LR.fit(X_train_scalar, Y_train)
model_scalar_LR_pred = model_scalar_LR_fit.predict(X_test_scalar)

r2_scalar_LR = r2_score(Y_test, model_scalar_LR_pred).round(4)
print("r2 score of Linear Regression_scalar data:", r2_scalar_LR)
mse_scalar_LR = mean_squared_error(Y_test, model_scalar_LR_pred).round(4)
print("mse error of Linear Regression_scalar data:", mse_scalar_LR)
mae_scalar_LR = mean_absolute_error(Y_test, model_scalar_LR_pred).round(4)
print("mae error of Linear Regression_scalar data:", mae_scalar_LR)
```

```python
model_scalar_RF = RandomForestRegressor(n_estimators=200)
model_scalar_RF_fit = model_scalar_RF.fit(X_train_scalar, Y_train.ravel())
model_scalar_RF_pred = model_scalar_RF_fit.predict(X_test_scalar)

r2_scalar_RF = r2_score(Y_test.ravel(), model_scalar_RF_pred).round(4)
print("r2 score of Random Forest Regression_scalar data:", r2_scalar_RF)
mse_scalar_RF = mean_squared_error(Y_test, model_scalar_RF_pred).round(4)
print("mse error of Random Forest Regression_scalar data:", mse_scalar_RF)
mae_scalar_RF = mean_absolute_error(Y_test, model_scalar_RF_pred).round(4)
print("mae error of Random Forest Regression_scalar data:", mae_scalar_RF)
```

```python
model_scalar_KNN = KNeighborsRegressor(n_neighbors=4)
model_scalar_KNN_fit = model_scalar_KNN.fit(X_train_scalar, Y_train)
model_scalar_KNN_pred = model_scalar_KNN_fit.predict(X_test_scalar)

r2_scalar_KNN = r2_score(Y_test, model_scalar_KNN_pred).round(4)
print("r2 score of KNeighborsRegressor_scalar data:", r2_scalar_KNN)
mse_scalar_KNN = mean_squared_error(Y_test, model_scalar_KNN_pred).round(4)
print("mse error of KNeighborsRegressor_scalar data:", mse_scalar_KNN)
mae_scalar_KNN = mean_absolute_error(Y_test, model_scalar_KNN_pred).round(4)
print("mae error of KNeighborsRegressor_scalar data:", mae_scalar_KNN)
```

### Plot Visualization

##### Creating dataframe for plot visualization

```python
model_name = ["LR", "LR", "RFR", "RFR", "KNN","KNN"]
data_type = ["Raw", "StandardScalar", "Raw", "StandardScalar", "Raw", "StandardScalar"]
r2_overall = [r2_LR, r2_scalar_LR, r2_RF, r2_scalar_RF, r2_KNN, r2_scalar_KNN]
mse_overall = [mse_LR, mse_scalar_LR, mse_RF, mse_scalar_RF, mse_KNN, mse_scalar_KNN]
mae_overall = [mae_LR, mae_scalar_LR, mae_RF, mae_scalar_RF, mae_KNN, mae_scalar_KNN]

data_part = {"Model":model_name, "Data_Type":data_type, "r2_score":r2_overall, "mse_error":mse_overall, "mae_error":mae_overall}
column_part = ["Model", "Data_Type", "r2_score", "mse_error", "mae_error"]

overall_part = pd.DataFrame(data_part, columns=column_part)
result_folder = "/content/drive/MyDrive/Data Analytics - Intern/Intern - Jayasri/Dataset/Energy Consumption in drone dataset/Result/"
overall_part.to_csv(result_folder+'R2_MSE_results.csv',index=False)
```

```python
fig = plt.figure(dpi=600)
ax = fig.add_subplot(111)
sns.barplot(data=overall_part, x='Model', y='r2_score', hue='Data_Type', edgecolor="black", linewidth=1.2)
ax.legend(fontsize=8, ncol=2, loc='upper right')
ax.set_title("model vs r2_score", fontsize=10, weight='bold')
plt.savefig(result_folder+"model_vs_r2_score.png", dpi=600)
```

```python
fig = plt.figure(dpi=600)
ax = fig.add_subplot(111)
sns.barplot(data=overall_part, x='Model', y='mse_error', hue='Data_Type', edgecolor="black", linewidth=1.2)
ax.legend(fontsize=8, ncol=2, loc='upper right')
ax.set_title("model vs mse_error", fontsize=10, weight='bold')
plt.savefig(result_folder+"model_vs_mse_error.png", dpi=600)
```

```python
fig = plt.figure(dpi=600)
ax = fig.add_subplot(111)
sns.barplot(data=overall_part, x='Model', y='mae_error', hue='Data_Type', edgecolor="black", linewidth=1.2)
ax.legend(fontsize=8, ncol=2, loc='upper right')
ax.set_title("model vs mae_error", fontsize=10, weight='bold')
plt.savefig(result_folder+"model_vs_mae_error.png", dpi=600)
```
