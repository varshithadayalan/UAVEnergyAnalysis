```python
import os
import numpy as np
import pandas as pd

# Data Preprocessing
from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler

# ML Model
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

# Performance Metrics
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

#Libraries for plots
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("flights.csv")
data
```

```python
data['altitude'] = data['altitude'].replace(['25','50','75','100'],[25,50,75,100])
```

```python
data = data[data.altitude.isin(['25-50-100-25','75']) == False]
```

```python
# Assuming you have a DataFrame named 'data' with columns: time, current, voltage, and c
# Create a new column 'c' with default values
data['C'] = 0
# Apply the condition using a loop
for index, row in data.iterrows():
    if row['time'] > 0:
        data.at[index, 'C'] = row['battery_current'] * row['battery_voltage'] * row['time']/3600
    else:
        data.at[index, 'C'] = row['battery_current'] * row['battery_voltage']/3600
```

```python
def heatmapcorr(Features, datatype):
    Feat_corr = Features
    print('Feature_Correlation:', Feat_corr)
    corr = data_scalar[Feat_corr].corr()
    corr_val = corr.sort_values(by='C',ascending=False).reset_index()
    #corr_val1 = corr_val.loc[(corr_val['C'] >= float(0.10))]
    #print(corr_val1)
    Feat_heatmap = list(corr_val['index'][1:11])
    print("Top_8_features:", Feat_heatmap)
    
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True
    with sns.axes_style("white"):
        f, ax = plt.subplots(figsize=(16, 8))
        ax = sns.heatmap(corr, mask=mask, cmap='RdYlGn',annot=True, edgecolor="black", linewidth=2)
        plt.xticks(fontsize=12, weight='bold')
        plt.yticks(fontsize=12, weight='bold')
        plt.title("{} Data".format(datatype),fontsize=18, weight='bold')
        #heatmap_plots = create_folder(result_folder, "Plots", "Heatmap_correlation")
        #plt.savefig(heatmap_plots+'Feature_importance_Density_{}.png'.format(datatype))
    return Feat_heatmap
```

```python
data_train= datar1.loc[(datar1['flight'] >= min(train_flightlist)) & (datar1['flight'] <= max(train_flightlist))]
data_test= datar1.loc[(datar1['flight'] >= min(test_flightlist)) & (datar1['flight'] <= max(test_flightlist))]
```

```python
x_train=data_train[Top_Comb_heatmap].values
y_train=data_train[['C']].values
x_test=data_test[Top_Comb_heatmap].values 
y_test=data_test[['C']].values

x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
y_train,y_test=y_train.ravel(), y_test.ravel()
```

```python
model_LR = LinearRegression()
model_LR_fit = model_LR.fit(x_train, y_train)
model_LR_pred = model_LR_fit.predict(x_test)

r2_LR = r2_score(y_test, model_LR_pred).round(4)
print("r2 score of Linear Regression:", r2_LR)
mse_LR = mean_squared_error(y_test, model_LR_pred).round(4)
print("mse error of Linear Regression:", mse_LR)
mae_LR = mean_absolute_error(y_test, model_LR_pred).round(4)
print("mae error of Linear Regression data:", mae_LR)
```

```python
model_RF = RandomForestRegressor()
model_RF_fit = model_RF.fit(x_train, y_train)
model_RF_pred = model_RF_fit.predict(x_test)

r2_RF = r2_score(y_test, model_RF_pred).round(4)
print("r2 score of Random Forest Regression:", r2_RF)
mse_RF = mean_squared_error(y_test, model_RF_pred).round(4)
print("mse error of Random Forest Regression:", mse_RF)
mae_RF = mean_absolute_error(y_test, model_RF_pred).round(4)
print("mae error of Random Forest Regression data:", mae_RF)
```

```python
model_KNN = KNeighborsRegressor(n_neighbors=10)
model_KNN_fit = model_KNN.fit(x_train, y_train)
model_KNN_pred = model_KNN_fit.predict(x_test)

r2_KNN = r2_score(y_test, model_KNN_pred).round(4)
print("r2 score of KNeighborsRegressor:", r2_KNN)
mse_KNN = mean_squared_error(y_test, model_KNN_pred).round(4)
print("mse error of KNeighborsRegressor:", mse_KNN)
mae_KNN = mean_absolute_error(y_test, model_KNN_pred).round(4)
print("mae error of KNeighborsRegressor data:", mae_KNN)
```
