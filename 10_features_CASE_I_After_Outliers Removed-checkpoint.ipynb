### Finding an outliers on batchwise and mass category

```python
def find_outliers_IQR(data_batch, mass):
    data_batch_mass = data_batch.loc[data_batch['mass'] == mass]
    print("Mass: ", mass, "kg")
    
    q1=data_batch_mass['Power'].quantile(0.25)
    q3=data_batch_mass['Power'].quantile(0.75)
#     print("Q1:", q1, "Q3:", q3)
    
    #Interquartile Range
    IQR=q3-q1
    
    Upper_fence = q3 + (1 * IQR)
    Lower_fence = q1 - (1 * IQR)
    print("Upper_Limit:", Upper_fence, "Lower_Limit:", Lower_fence)
    
    outliers = data_batch_mass['Power'][((data_batch_mass['Power']<Lower_fence) | (data_batch_mass['Power']>Upper_fence))]
    outliers = outliers.reset_index()
    print("Number of outliers:", len(outliers["index"]))
    outliers_dropped = data_batch_mass.drop(outliers['index'])
    print("Before Outlier: ", data_batch_mass.shape,"After_Outlier", outliers_dropped.shape)
    print("----------------------------------------------------------------------")
    return outliers_dropped
```

```python
# regressor = SVR(kernel='linear')
regressor = SVR(kernel='linear')
regressor.fit(X_train_scale, y_train.ravel())

result = permutation_importance(regressor, X_train_scale, y_train.ravel(), scoring='neg_mean_squared_error')
sorted_indices = result.importances_mean.argsort()[::-1]

print("Feature importance:")
for i in sorted_indices:
    feature_name = X_train.columns[i]
    score = result.importances_mean[i]
    print(f"{feature_name}: {score}")

svr_importance_scores = result.importances_mean[sorted_indices]
names = X_train.columns[sorted_indices]

plt.bar(names, svr_importance_scores)
plt.xticks(rotation=90)
plt.title("Permutation Feature Importance")
plt.xlabel("Feature")
plt.ylabel("SVR Feature Importance")
plt.show()
```

### K-Nearest Neighbours

```python
regressor = KNeighborsRegressor(n_neighbors=1)
regressor.fit(X_train_scale, y_train.ravel())

result = permutation_importance(regressor, X_train_scale, y_train.ravel(), n_repeats=10, random_state=42, n_jobs=-1)
sorted_indices = result.importances_mean.argsort()[::-1]

knn_importance_scores = result.importances_mean[sorted_indices]
names = X_train.columns[sorted_indices]
knn_score_per = (knn_importance_scores*100).round(2)
print(knn_score_per)
```

```python
knn_score_final = []
for i in range(0, len(knn_score_per)):
    list_per = ((knn_score_per[i]/sum(knn_score_per))*100).round(2)
    knn_score_final.append(list_per)
```

```python
feature_importance_knn = pd.DataFrame({'Feature': X_train.columns, 'Importance': knn_score_final})
feature_importance_knn = feature_importance_knn.sort_values('Importance', ascending=False)
```

```python
plt.figure(dpi=600)
# plt.bar(feature_importance_df['Feature'], (feature_importance_df['Importance']*100).round(2), color='b')
ax = sns.barplot(x="Importance", y="Feature", data=feature_importance_knn, color="b")
ax.text(11, 12, "K-Nearest Neighbors", 
        fontsize = 15,          # Size
        fontstyle = "oblique",  # Style
        fontweight= 'bold',
        color = "red",          # Color
        ha = "center", # Horizontal alignment
        va = "center") # Vertical alignment 
ax.set(xlabel=None, ylabel=None)
plt.xticks(rotation=0)
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')
plt.savefig("Feature_Importance_KNN_horiz.png", dpi=600)
plt.show()
```
