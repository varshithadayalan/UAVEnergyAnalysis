```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'data' containing the values

# Calculate the average of each column
energy8 = train8['energyh'].mean()
energy170=train170['energyh'].mean()
energy108=train108['energyh'].mean()
energy124=train124['energyh'].mean()
energy146=train146['energyh'].mean()
energy177=train177['energyh'].mean()
energy77=train77['energyh'].mean()
energy116=train116['energyh'].mean()
energy10=test10['energyh'].mean()
energy203=test203['energyh'].mean()
energy12=test12['energyh'].mean()
energy171=test171['energyh'].mean()

avg_values=[energy8,energy170,energy108,energy124,energy146,energy177,energy116,energy77,energy10,energy203,energy12,energy171]

# Create a bar plot using the average values
plt.bar(['energy8','energy170','energy108','energy124','energy146','energy177','energy116','energy77','energy10','203','energy12','energy171'],avg_values )

# Set the labels and title
plt.xlabel('Flights of P0,A25 and varying speed')
plt.ylabel('Average')
plt.title('Average Values of Columns')

# Rotate the x-axis labels if needed
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
combined_train = pd.concat([train8, train170, train108, train124, train146, train177, train77, train116],ignore_index=True) 
combined_train
```

```python
combined_test=pd.concat([test10, test203, test12, test171],ignore_index=True)
combined_test
```

```python
x_train=combined_train[['time', 'wind_speed', 'battery_voltage', 'battery_current',
       'position_x', 'position_y', 'position_z', 'orientation_y', 'velocity_z',
       'altitude']].values

y_train=combined_train[['energyh']].values

x_test=combined_test[['time', 'wind_speed', 'battery_voltage', 'battery_current',
       'position_x', 'position_y', 'position_z', 'orientation_y', 'velocity_z',
       'altitude']].values

y_test=combined_test[['energyh']].values

x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
model_LR = LinearRegression()
model_LR_fit = model_LR.fit(x_train, y_train.ravel())
model_LR_pred = model_LR_fit.predict(x_test)

r2_LR = r2_score(y_test.ravel(), model_LR_pred).round(4)
print("r2 score of Linear Regression:", r2_LR)
mse_LR = mean_squared_error(y_test.ravel(), model_LR_pred).round(4)
print("mse error of Linear Regression:", mse_LR)
mae_LR = mean_absolute_error(y_test.ravel(), model_LR_pred).round(4)
print("mae error of Linear Regression data:", mae_LR)
```

```python
print("predicted value:",model_LR_pred)
print("actual value:",y_test.ravel())
```

```python
model_KNN = KNeighborsRegressor(n_neighbors=5)
model_KNN_fit = model_KNN.fit(x_train, y_train)
model_KNN_pred = model_KNN_fit.predict(x_test)

r2_KNN = r2_score(y_test, model_KNN_pred).round(4)
print("r2 score of KNeighborsRegressor:", r2_KNN)
mse_KNN = mean_squared_error(y_test, model_KNN_pred).round(4)
print("mse error of KNeighborsRegressor:", mse_KNN)
mae_KNN = mean_absolute_error(y_test, model_KNN_pred).round(4)
print("mae error of KNeighborsRegressor data:", mae_KNN)
print("Y test's actual values: ",y_test)
print("Predicted value: ",model_KNN_pred)
```

```python
from scipy import stats

z_scores = stats.zscore(data['time'])
threshold = 3  # Adjust as needed
outliers = (z_scores > threshold) | (z_scores < -threshold)
outlier_indices = data.index[outliers]
```

```python
Q1 = data['time'].quantile(0.25)
Q3 = data['time'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = (data['time'] < lower_bound) | (data['time'] > upper_bound)
outlier_indices = data.index[outliers]
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(contamination=0.05)  # Adjust contamination parameter
model.fit(data[['time']])
outliers = model.predict(data[['time']]) == -1
outlier_indices = data.index[outliers]
```

```python
plt.scatter(data.index, data['time'], c='blue', label='Inliers')
plt.scatter(outlier_indices, data.loc[outlier_indices, 'time'], c='red', label='Outliers')
plt.legend()
plt.show()
```

```python
# Remove outliers from the dataset
data_cleaned = data.drop(outlier_indices)

# Alternatively, you can keep only the inliers
data_inliers = data[~data.index.isin(outlier_indices)]
```
