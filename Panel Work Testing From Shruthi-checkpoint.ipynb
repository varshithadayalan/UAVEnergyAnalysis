```python
get_ipython().system('pip install seaborn')
get_ipython().system('pip install matplotlib')
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
flights=pd.read_csv('flights.csv')
flights
flights = flights[flights['altitude'] != '25-50-100-25']
flights
flights['altitude'] = pd.to_numeric(flights['altitude'], errors='coerce')
flights.info()
sns.heatmap(flights.corr(numeric_only=True),cbar=True,linewidths=1)

```

```python
flights['power']=flights['battery_voltage']*flights['battery_current']
```

```python
get_ipython().system('pip install panel')
import panel as pn 
pn.extension('tabulator')
import hvplot.pandas
import plotly.express as px
```

```python
def plot_altitude(altitude):
    df = flights[flights['altitude'] == altitude]
    fig = px.scatter(df, x='time', y='power', color_discrete_sequence=['orange'])
    fig.update_layout(title=f"Power vs Time for Altitude {altitude}")
    return fig
```

```python
#MAIN DASHBOARD PART 
# create a function to generate a plot for a given altitude
def plot_altitude(altitude):
    df = flights[flights['altitude'] == altitude]
    fig = px.scatter(df, x='time', y='power', color_discrete_sequence=['orange'])
    fig.update_layout(title=f"Power vs Time for Altitude {altitude}")
    return fig

# create a radio button for altitude choices
altitudes = [0, 3, 25, 50, 75, 100]
title = pn.pane.Markdown("<b>Change the Altitude</b>", style={"font-size": "24px","color": "#FF8427","background":"#FFFFFF"})

altitude_buttons1 = pn.widgets.RadioButtonGroup(
    options=altitudes, value=0, button_type='warning', inline=True)
altitude_buttons1.background = '#F9E79F'
altitude_buttons1.text_color = 'blue'
altitude_buttons1.font_size = '16pt'
altitude_buttons1.font_weight = 'bold'
altitude_buttons1.border = '2px solid black'
altitude_buttons1.margin = (10, 20, 10, 20)
altitude_buttons1.padding = 10
alti_butts=pn.Column(title,altitude_buttons1,align='center')
# create a plot for the initial altitude
initial_altitude = altitude_buttons1.value
initial_plot = plot_altitude(initial_altitude)
plot_panel = pn.panel(initial_plot)

# create a function to update the plot when the altitude button is changed
def update_plot_alti(event):
    altitude = event.new
    plot = plot_altitude(altitude)
    plot_panel.object = plot

# add the altitude button and plot to a column
plot_column = pn.Column(
    alti_butts,
    plot_panel,
    sizing_mode='fixed', height=500,
    width=1000,
    align='center'
)
altitude_buttons1.height = 30
# register the update function to the altitude button
altitude_buttons1.param.watch(update_plot_alti, 'value')
# create a paragraph on the left side
dashboard_text = """
# Flight Data Dashboard

This dashboard shows the power consumption of an Unmanned Aerial Vehicle(DJI MATRICE 100) for various attributes such as time and route and for altitudes and payloads. The sample dataset used here is and some of the variables used are \n 
<b>Route
       Description:</b> A predefined path followed by the aircraft. The differences among routes R1 to R7 reflect variations on the starting point or variations on the altitude during cruise. \n   
R1 to R7 = full flights completing a cruise movement; \n
A1 = Ancillary ground test with no propellers and no motor movement;\n
A2 = Ancillary ground test with no propellers and minimum movement; \n
A3 = Ancillary ground test with propellers and minimum movement;\n
H = Hover test with no horizontal movement. \n
   <b> Power:</b> The Power consumed by the drone at a particular instant of time. \n
   <b>Altitude:</b>Predefined altitude in meters (m) (25 m, 50 m, 75 m and 100 m). The aircraft takes off vertically until it reaches the preset altitude
"""

dashboard_panel = pn.pane.Markdown(dashboard_text, sizing_mode='fixed', width=400,style={"font-size":"16px","background":"#FFC800"})



def plot_power_route(altitude):
    if altitude is not None:
        df = flights[flights['altitude'] == altitude]
        title = f"Power vs Route for Altitude {altitude}"
    else:
        df = flights
        title = "Power vs Route for All Altitudes"
   
    fig = px.scatter(df, x='route', y='power', color='time',category_orders={'route':["A1","A2","A3","R1","R2","R3","R4","R5","R6","H"]})
    fig.update_layout(title=title)
    return fig
title2 = pn.pane.Markdown("<b>Change the Altitude</b>", style={"font-size": "24px","color": "#FF8427","background":"#FFFFFF"})

altitude_buttons2 = pn.widgets.RadioButtonGroup(options=altitudes, value=None, button_type='warning', inline=True)
altitude_buttons2.background = '#F9E79F'
altitude_buttons2.text_color = 'blue'
altitude_buttons2.font_size = '16pt'
altitude_buttons2.font_weight = 'bold'
altitude_buttons2.border = '2px solid black'
altitude_buttons2.margin = (10, 20, 10, 20)
altitude_buttons2.padding = 10
plot_options = ['Single Altitude', 'All Altitudes']
plot_buttons = pn.widgets.RadioButtonGroup(options=plot_options, value='Single Altitude', button_type='warning', inline=True)
plot_buttons.background = '#F9E79F'
plot_buttons.text_color = 'blue'
plot_buttons.font_size = '16pt'
plot_buttons.font_weight = 'bold'
plot_buttons.border = '2px solid black'
plot_buttons.margin = (10, 20, 10, 20)
plot_buttons.padding = 10
alti2_butts=pn.Column(title,altitude_buttons2,plot_buttons,align='center')


initial_plot_route = plot_power_route(None)
plot_route=pn.panel(initial_plot_route)
# create a function to update the plot when the altitude button is changed

    
def update_plot(event):
    if plot_buttons.value == 'Single Altitude':
        altitude = event.new
        plot = plot_power_route(altitude)
    else:
        plot = plot_power_route()
    plot_route.object = plot

altitude_buttons2.param.watch(update_plot, 'value')
plot_buttons.param.watch(update_plot, 'value')

# add the altitude and plot selection buttons and plot to a panel
plot_some = pn.Column(
    alti2_butts,
    plot_route,
    sizing_mode='fixed',width=1000,align='center'
)
vspacer = pn.layout.VSpacer(height=50)

hoho=pn.Column(plot_column,vspacer,plot_some,sizing_mode='fixed')

# register the update function to the altitude button
# add the altitude buttons and plot to a row
row = pn.Row(
    dashboard_panel,
    hoho,
    sizing_mode='stretch_both',
    background="#004E89"
)



# display the row
row.show()
```

```python
import panel as pn
import matplotlib.pyplot as plt
import plotly.graph_objs as go
from panel.widgets import Select
pn.extension("plotly")

options = flights['flight'].astype(str).tolist()
options=list(set(options))
flight_button = pn.widgets.Select(name="Select a suitable flight number", options=options)

def plottingsome(flight_number):
    # Matplotlib
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Plotly
    fig_plotly = go.Figure(data=[go.Scatter3d()])

    flights_filtered = flights[flights['flight'] == flight_number]

    x = flights_filtered['position_x']
    y = flights_filtered['position_y']
    z = flights_filtered['position_z']
    yaw = flights_filtered['orientation_x']
    pitch = flights_filtered['orientation_y']
    roll = flights_filtered['orientation_z']

    ax.plot(x, y, z)

    # Plotly
    fig_plotly.add_trace(go.Scatter3d(x=x, y=y, z=z, mode='lines'))

    # Matplotlib
    ax.scatter(x, y, z)

    # Plotly
    fig_plotly.add_trace(go.Scatter3d(x=x, y=y, z=z, mode='markers'))

    # Matplotlib
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Drone Position and Orientation')

    # Plotly
    fig_plotly.update_layout(scene=dict(xaxis_title='X-values', yaxis_title='Y-values', zaxis_title='Z-values'), title='Drone Position and Orientation')
    return fig_plotly

flight_number = flight_button.value
initial_plot = plottingsome(flight_number)
plot_panel = pn.pane.Plotly(sizing_mode='stretch_both', config={'responsive': True})

dashboard_text = """
# Flight Data Dashboard

This dashboard shows the power consumption of an Unmanned Aerial Vehicle(DJI MATRICE 100) for various attributes such as time and route and for altitudes and payloads. The sample dataset used here is and some of the variables used are \n 
<b>Route
       Description:</b> A predefined path followed by the aircraft. The differences among routes R1 to R7 reflect variations on the starting point or variations on the altitude during cruise. \n   
R1 to R7 = full flights completing a cruise movement; \n
A1 == Ancillary ground test with no propellers and no motor movement;\n
A2 = Ancillary ground test with no propellers and minimum movement; \n
A3 = Ancillary ground test with propellers and minimum movement;\n
H = Hover test with no horizontal movement. \n
   <b> Power:</b> The Power consumed by the drone at a particular instant of time. \n
   <b>Altitude:</b>Predefined altitude in meters (m) (25 m, 50 m, 75 m and 100 m). The aircraft takes off vertically until it reaches the preset altitude
"""

def on_click(event):
    flight_number=event.new
    plot=plottingsome(flight_number)
    plot_panel.object=plot
    

dashboard_panel = pn.pane.Markdown(dashboard_text, sizing_mode='fixed', width=400,style={"font-size":"16px","background":"#FFC800"})

row=pn.Row(dashboard_panel,plot_panel,flight_button)
row.show()
```
