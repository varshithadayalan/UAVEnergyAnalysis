```python
import pandas as pd

# Define a function to convert columns to numeric except for specific columns
def convert_columns_to_numeric(df):
    # Get a list of columns to exclude from conversion
    exclude_columns = ['time', 'date', 'route']
    
    # Iterate through columns and convert to numeric if not in exclude list
    for column in df.columns:
        if column not in exclude_columns:
            df[column] = pd.to_numeric(df[column], errors='coerce')    
    return df

```
